# **Predicting Wind Turbine Power Generation Using Matchine Learning**

#Objective
* We are predicting the normalized power output of a wind turbine.
* The model predicts the turbine's efficiency/production level, normalized between 0 and 1.
* Our objective is power forecasting/prediction in a normalized range.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Loading the dataset
df = pd.read_csv('/content/data.csv')
```

```python
# Data Cleaning (Handle missing values)
df = df.dropna()
```

```python
# Selecting features and target
features = ['temperature_2m', 'relativehumidity_2m','dewpoint_2m', 'windspeed_10m', 'windspeed_100m', 'winddirection_10m', 'winddirection_100m','windgusts_10m']
target = ['Power']
X = df[features]
y = df[target]
```

Splitting the dataset into 80% and 20%
1. 80% training to build the model
2. 20% testing to evaluate it

```python
train_X, val_X, train_y, val_y = train_test_split(X, y,test_size=0.2, random_state = 2)
```

```python
# Initializing and training the model
forest_model = RandomForestRegressor(random_state=2)
forest_model.fit(train_X, train_y)
```

```python
# Making Prediction
predictions = forest_model.predict(val_X)
```

```python
#Evaluating the model
mse = mean_absolute_error(val_y,predictions)
r2 = r2_score(val_y,predictions)
accuracy_percentage = r2 * 100
print(predictions)
print(f"Mean Absolute Error: {mse}")
print(f"R-squared Score: {r2}")
print(f"Accuracy: {accuracy_percentage:.2f}%")
```

Accuracy for this model is coming as 73.31%
